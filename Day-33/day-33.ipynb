### Task 2: Correlation Practice 
1. Create dataset 
2. Find correlation matrix 
3. Analyze relationships 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create dataset
data = pd.DataFrame({
    "StudyHours": [2,4,6,8,10],
    "Marks": [40,50,65,80,95],
    "SleepHours": [9,8,7,6,5]
})
```

```python
corr_matrix = data.corr()
print(corr_matrix)
```

- StudyHours vs Marks -- Strong positive correlation
- StudyHours vs Sleep -- Strong negative correlation
- Marks vs Sleep -- Strong negative correlation

### Task 3: Heatmap Practice 
1. Create heatmap 
2. Show annotations 
3. Analyze strong correlations 

```python
# Create heatmap 
sns.heatmap(corr_matrix)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#show annotations
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Heatmap with Values")
plt.show()
```

- StudyHours vs Marks -- Strong positive correlation
- StudyHours vs Sleep -- Strong negative correlation
- Marks vs Sleep -- Strong negative correlation

### Task 4: Pairplot Practice 
1. Create pairplot 
2. Analyze distributions 
3. Compare variables 

```python
#create pairplot
sns.pairplot(data)
plt.show()
```


#### 1. StudyHours vs Marks
- Strong **positive correlation**
- As StudyHours increases → Marks increase
- Clear **upward linear trend**

#### 2. StudyHours vs SleepHours
- Strong **negative correlation**
- More StudyHours -- less SleepHours
- Clear **downward trend**


#### 3. Marks vs SleepHours
- Strong **negative correlation**
- Higher Marks --lower SleepHours
- Indicates inverse relationship

### Task 5: Scatter Plot Practice 
1. Study hours vs marks 
2. Salary vs experience 
3. Temperature vs sales 

```python
# Study hours vs marks
sns.scatterplot(x="StudyHours", y="Marks", data=data)
plt.title("Study Hours vs Marks")
plt.show()
```

```python
salary_data = pd.DataFrame({"Experience":[1,2,3,4,5], "Salary":[30000,40000,50000,60000,70000]})
sns.scatterplot(x="Experience", y="Salary", data=salary_data)
plt.title("Salary vs Experience")
plt.show()
```

```python
# Temperature vs sales
temp_sales = pd.DataFrame({"Temperature":[20,25,30,35,40], "Sales":[100,150,200,250,300]})
sns.scatterplot(x="Temperature", y="Sales", data=temp_sales)
plt.title("Temperature vs Sales")
plt.show()
```
