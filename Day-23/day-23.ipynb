3. Divide array by scalar 

```python
# add scalar to array
import numpy as np
a = np.array([11,22,33,44,55])
print("add :", a+10)
print("multiply :", a*10)
print("divide :", a/11)
```

### Task 3: Array Broadcasting 
1. Add two compatible arrays 
2. Multiply arrays 
3. Observe shape changes 

```python
b = np.array([[1,2],[3,4],[5,6]])
c = np.array([[2],[6],[8]])
print("b's shape:", b.shape)
print("c's shape:", c.shape)
add_bc = b+c
multiply_bc = b*c
print("addition: ",add_bc,"\nshape:", add_bc.shape)
print("multiplication:",multiply_bc,"\nshape:", multiply_bc.shape)
```

### Task 4: Shape Practice 
1. Print shapes 
2. Check compatible shapes 
3. Try incompatible shapes 

```python
#compatible dimensions
b = np.array([[1,2,3],[4,5,6]])
c = np.array([2,6,8])
print(b.shape)
print(c.shape)
add_bc = b+c
multiply_bc = b*c
print("addition: ",add_bc,"\nshape:",add_bc.shape)
print("multiplication:",multiply_bc,"\nshape:",multiply_bc.shape)
```

```python
#incompatible dimensions
b = np.array([[1,2,3],[4,5,6]])
c = np.array([2,6,8,9])
print(b+c)
print(b*c)

```

### Task 5: 2D Broadcasting 
1. Add scalar to matrix 
2. Multiply matrix 
3. Normalize matrix values 

```python
arr_1 = np.array([[10,20,30],[40,50,60]])
print("addition:", arr_1 + 4)
print("multiplication:", arr_1*10)
```

```python
normalized = (arr_1 - arr_1.min()) / (arr_1.max() - arr_1.min())

print("\nNormalized Matrix:\n", normalized)
```
