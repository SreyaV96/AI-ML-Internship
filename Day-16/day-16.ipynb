### Task 2: Cleaning Practice 
1. Remove None values 
2. Remove negative values 
3. Keep only valid data 

```python
students = [{"Name":"Anagha", "Marks":80,"Grade":"A"},
            {"Name":"Anjitha", "Marks":76,"Grade":"B"},
            {"Name":"Anusha", "Marks":40,"Grade":"D"},
            {"Name":"Heera", "Marks":36,"Grade":"E"},
            {"Name":"Manasa","Marks":None, "Grade":None},
            {"Name":"Geethu","Marks":-45, "Grade":"C"},
            {"Name":"Taniya","Marks":120, "Grade":"D"}]
```

```python
#using list comprehension
cleaned = [s for s in students if s["Marks"] is not None and 0 <= s["Marks"] <= 100]
print(cleaned)
```

```python
cleaned =[]
for s in students: 
    if s["Marks"] is not None and 0<=s["Marks"] <=100: 
        cleaned.append(s) 
print(cleaned) 
```

### Task 3: Transformation 
1. Add pass/fail field 
2. Convert marks to percentage 
3. Create new calculated field 

```python
for s in cleaned:
    s["status"] = "Pass" if s["Marks"] >= 45 else "Fail"
```

```python
print(cleaned)
```

```python
total_marks =100
for s in cleaned:
    s["percentage"] = (s["Marks"]/total_marks)*100
print(cleaned)
```

### Task 4: Sorting 
1. Sort students by marks 
2. Print top 3 students 

```python
cleaned.sort(key=lambda x: x["Marks"],reverse=True)
print(cleaned)
```

```python
print("Top 3 Students:")
for student in cleaned[:3]:
    print(student["Name"], student["Marks"])
```

### Task 5: Grouping 
1. Group pass and fail 
2. Count each group 

```python
#group pass and fail
result ={"Pass":[], "Fail": []}
for s in cleaned:
    if s["Marks"] >= 45:
        result["Pass"].append(s)
    else:
        result["Fail"].append(s)
print(result)
```

```python
print("Pass count: ", len(result["Pass"]))
print("Fail count : ", len(result["Fail"]))
```

### Task 6: Data Analysis 
1. Find average marks 
2. Find highest marks 
3. Find lowest marks 

```python
#Average marks
total = sum(s["Marks"] for s in cleaned)
avg = total / len(cleaned)
print("Average Marks : ", avg)
```

```python
#highest and lowest marks
highest = max(s["Marks"] for s in cleaned)
lowest = min(s["Marks"] for s in cleaned)
print("Highest Marks: ",highest)
print("Lowest Marks: ", lowest)
```
