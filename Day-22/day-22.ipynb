### Task 2: Basic Aggregation 
1. Find sum 
2. Find mean 
3. Find max and min 

```python
import numpy as np
a = np.array([12,14,16,18,20])
print("sum:", np.sum(a))
print("mean:", np.mean(a))
print("max:", np.max(a))
print("min:", np.min(a))
```

### Task 3: Statistical Functions 
1. Find median 
2. Find standard deviation 
3. Find variance 

```python
arr_1 = np.array([49,56,43,78,45,89])
print("median:", np.median(arr_1))
print("std dev:", np.std(arr_1))
print("varriance:", np.var(arr_1))
```

### Task 4: Axis Practice 
1. Row-wise sum 
2. Column-wise sum 
3. Mean using axis 

```python
arr_2 = np.array([[2,4,6],[8,10,12],[14,16,18]])
print("row wise sum:", np.sum(arr_2, axis=1))
print("column wise sum:", np.sum(arr_2, axis=0))
print("row wise mean:", np.mean(arr_2, axis=1))
print("column wise mean:" ,np.mean(arr_2, axis=0))
```

### Task 5: Index Functions 
1. Find index of max 
2. Find index of min 

```python
arr_3 = [23,31,26,45,12]
print("index of max:", np.argmax(arr_3))
print("index of min:", np.argmin(arr_3))
```

### Task 6: Cumulative 
1. Compute cumulative sum 
2. Analyze output 

```python
b = np.array([6,2,4,7,9,3,10])
print("cumulative sum:", np.cumsum(b))
print("Analysis:")
print("Each element is the running total of previous elements.")
```
