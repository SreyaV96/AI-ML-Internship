### Task 2: Sorting Practice 
1. Sort salary ascending 
2. Sort salary descending 
3. Sort by multiple columns

```python
import pandas as pd
#create dataset
data ={"Name":["Arunima","Akhil","Deepika","Dona","Evon","Sagar","Meera","Hari","Sathya","Anagha"],
        "Salary":[20000,60000,40000,30000,45000,35000,15000,80000,55000,35000],
        "Department":["HR","AI","QA","DBA","AI","DBA","HR","AI","DBA","QA"],
        "Category":["a","c","b","c","b","b","b","a","a","b"],
        "City":["Trivandrum","Kochi","Bangalore","Calicut","Trivandrum","Kochi","Bangalore","Kochi","Calicut","Trivandrum"]}
df = pd.DataFrame(data)
df
```

```python
#sort salary ascending
df.sort_values("Salary")
```

```python
#sort salary descending
df.sort_values("Salary",ascending=False)
```

```python
#sort by multiple columns
df.sort_values(["Department","Salary"],ascending=[True,True])
```

### Task 4: Aggregation Practice 
1. Find average salary 
2. Find total salary 
3. Count employees 

```python
# average salary
print(df["Salary"].mean())
```

```python
#average salary by department
df.groupby("Department")["Salary"].mean()
```

```python
# total salary
print("Total Salary:", df["Salary"].sum())
```

```python
#total salary by department
df.groupby("Department")["Salary"].sum()
```

```python
#count employees 
print(df["Name"].count())
```

```python
#employees per department
df.groupby("Department")["Name"].count()
```
