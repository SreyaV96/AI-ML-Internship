## Task 2: Function Practice 

### 1. Function to find factorial 

```python
def factorial(n):
    if n==1 :
        return 1
    return n*factorial(n-1)
```

```python
print(factorial(5))
```

### 2. Function to check prime number 

```python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True
```

```python
print(is_prime(7))
```

```python
print(is_prime(6))
```

### 3. Function to reverse string 

```python
def reverse_str(text):
    return text[::-1]
```

```python
print(reverse_str("python"))
```

### 4. Function to calculate average 

```python
def avg(nums):
    return sum(nums)/len(nums)
```

```python
print(avg([10, 20, 30]))
```

## Task 3: Recursion Practice 

### 1. Factorial using recursion 

```python
def factorial(n):
    if n==1 :
        return 1
    return n*factorial(n-1)
```

```python
factorial(3)
```

### 2. Sum of numbers using recursion

```python
def sum_numbers(n):
    if n == 1:
        return 1
    return n + sum_numbers(n - 1)
```

```python
n = int(input("Enter a positive number: "))
print("Sum =", sum_numbers(n))
```

### 3. Fibonacci series 

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

terms = int(input("Enter number of terms: "))

print("Fibonacci Series:")

for i in range(terms):
    print(fibonacci(i), end=" ")
```

## Task 4: Lambda & Map 

### 1. Square all numbers 

```python
nums = [1, 2, 3, 4] 
squares = list(map(lambda x: x*x, nums)) 
print(squares) 
```

### 2. Double all values 

```python
result = list(map(lambda x: x*2, nums)) 
print(result) 
```

### 3. Convert list to uppercase 

```python
words = ["apple", "grapes","banana", "mango"]
result = list(map(lambda x: x.upper(), words))
print("Uppercase List:", result)
```

## Task 5: Filter Practice 

### 1. Find even numbers 

```python
nums = [1, 2, 3, 4] 
result = list(filter(lambda x: x % 2 == 0, nums)) 
print("even numbers :", result) 
```

### 2. Filter students with marks > 50 

```python
marks = [45, 67, 89, 32, 50, 76, 90]
result = list(filter(lambda x: x > 50, marks))
print("Marks Greater than 50:", result)
```

### 3. Filter words with length > 5 

```python
words = ["apple", "grapes","banana", "mango","kiwi", "guava"]
result = list(filter(lambda x: len(x) > 5, words))
print("Words with Length > 5:", result)
```
