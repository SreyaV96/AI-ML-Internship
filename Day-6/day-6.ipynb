### 4. Count even and odd numbers

```python
numbers = [2,1,4,3,6,5,9,11,12]
even = 0
odd = 0
for num in numbers:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

## Task 3: String Problems 

### 1. Reverse string 

```python
s = input ("Enter a text")
print(s[::-1])
```

### 2. Check palindrome

```python
s = input("Enter a text")
if s == s[::-1]:
    print("Palindrome")
else :
    print("not a palindrome")
```

### 3. Count characters 

```python
text = "Malayalam" 
count = {} 
for ch in text: 
    count[ch] = count.get(ch, 0) + 1 
print(count) 
```

### 4. Count vowels and consonants 

```python
a = input("Enter a text: ").lower()
vowels = "aeiou"
count_v = 0
count_c = 0
for ch in a:
    if ch.isalpha():  # check only letters
        if ch in vowels:
            count_v += 1
        else:
            count_c += 1

print("Number of vowels:", count_v)
print("Number of consonants:", count_c)
```

## Task 4: Dictionary Problems 

### 1. Create student list 

```python
students = [{"Name":"Anand", "Marks":69},
            {"Name":"Sarah", "Marks":87},
            {"Name":"Meenu","Marks":45},
            {"Name":"Vishnu","Marks":58},
            {"Name":"Sreelakshmi","Marks":90}]
```

### 2. Find topper 

```python
top = students[0] 
for s in students: 
    if s["Marks"] > top["Marks"]:
        top = s 
print(top) 
```

### 3. Calculate average

```python
marks = [student["Marks"] for student in students]
average = sum(marks) / len(marks)
print(average)
```

### 4. Filter passed students

```python
passed_students = [student for student in students if student["Marks"] >= 50]
print(passed_students)
```

### 1. Find second smallest number

### 2. Merge two lists without duplicates

```python
l1 = [1,2,3,4,8,9,10,12,1,12,13]
l2 = [5,10,15,20,25,30,25,10,35,45,35]
merged = list(set(l1+l2))
print(merged)
```

### 3. Find frequency of elements 

```python
l2 = [5,10,15,20,25,30,25,10,35,45,35]
freq = {}
for n in l2:
    freq[n] = freq.get(n, 0) + 1

print(freq)
```
