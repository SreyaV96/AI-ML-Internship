### Task 1: Theory 
1. What is Machine Learning? 
- Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn from data and improve their performance without being explicitly programmed.
- ML systems identify patterns in data and use those patterns to make predictions or decisions.
- Example:
    - A music app learning your favorite songs.
    - A shopping website suggesting products you may like

---
2. Why ML is important? 
- Machine Learning is important because it helps computers:
 - Automate tasks
 - Make accurate predictions
 - Analyze large amounts of data quickly
 - Improve user experience
 - Reduce human effort
- Importance in Real Life
 - Healthcare: Disease prediction
 - Banking: Fraud detection
 - Education: Personalized learning
 - Business: Customer analysis
 - Entertainment: Movie recommendations

---

3. What is dataset? 
- A dataset is a structured collection of data used to train and evaluate ML models. It usually consists of:
 - Features (inputs): The variables or attributes used for learning.
 - Labels (outputs): The target values the model tries to predict.
- Examples (rows): Each record containing features and labels.

---
4. Difference between AI and ML 

| AI (Artificial Intelligence) | ML (Machine Learning)          |
| ---------------------------- | ------------------------------ |
| Makes machines intelligent   | Helps machines learn from data |
| Broad field                  | Subset of AI                   |
| Uses logic and rules         | Uses data and algorithms       |
| Example: Virtual assistants  | Example: Spam filter           |

---

5. Types of Machine Learning 
 1. Supervised Learning : Model learns from labeled data (e.g., predicting house prices).

 2. Unsupervised Learning : Model finds hidden patterns in unlabeled data (e.g., customer segmentation).

 3. Reinforcement Learning : Model learns by interacting with an environment and receiving rewards or penalties (e.g., training robots or game AI).

```python
# Customer grouping
data = {
    "Age": [18, 22, 25, 45, 50, 48],
    "Spending": [2000, 2500, 3000, 8000, 8500, 9000]
}

df = pd.DataFrame(data)
df
```

- Similar customers fall into the same cluster.

```python
#pattern finding
transactions = [
    ["Bread", "Butter"],
    ["Milk", "Bread"],
    ["Bread", "Butter", "Jam"],
    ["Milk", "Bread", "Butter"]
]

# Find transactions containing Bread and Butter
count = 0

for item in transactions:
    if "Bread" in item and "Butter" in item:
        count += 1

print("Bread and Butter bought together:", count, "times")
```

- The program finds shopping patterns.
- Customers buying Bread often buy Butter.

### Task 5: Reinforcement Learning Practice 
1. Robot learning example 
2. AI game example 
3. Reward/penalty explanation 

```python
# Simple Robot Reward System

position = 0

# Robot moves
moves = ["right", "right", "wrong", "right"]

reward = 0

for move in moves:

    if move == "right":
        position += 1
        reward += 10
        print("Correct Move : Reward +10")

    else:
        reward -= 5
        print("Wrong Move : Penalty -5")

print("Final Position:", position)
print("Total Reward:", reward)
```

- Correct movement gives reward.
- Wrong movement gives penalty.
- Robot learns better actions over time.

##### Reward/penaltiy system
- Good action - Reward added
- Bad action - Penalty applied
- The agent tries to increase total rewards by choosing better actions.

```python
# Simple Game AI Reward Example

game_results = ["win", "lose", "win", "win"]

score = 0

for result in game_results:

    if result == "win":
        score += 20
        print("Win → Reward +20")

    else:
        score -= 10
        print("Lose → Penalty -10")

print("Final Score:", score)
```

- The system:
    - Gets rewards for winning
    - Gets penalties for losing
- Learns strategies to maximize rewards
