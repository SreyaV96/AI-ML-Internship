### Task 2: Dataset Practice 
1. Create student dataset 
2. Create salary dataset 
3. Create sales dataset 

```python
import pandas as pd
#student dataset
student_data = {
    'Hours_Studied': [2, 4, 6, 8, 10],
    'Attendance': [60, 70, 80, 90, 95],
    'Marks': [50, 68, 75, 82, 90]
}

student_df = pd.DataFrame(student_data)
print(student_df)
```

```python
#salary dataset
salary_df = pd.DataFrame({
    'Experience': [1, 2, 3, 4, 5],
    'Salary': [25000, 30000, 40000, 50000, 60000]
})

print(salary_df)
```

```python
#sales data
sales_data = {
    'Advertising_Spend': [1000, 2000, 3000, 4000, 5000],
    'Salespeople': [5, 10, 15, 20, 25],
    'Product_Price': [50, 45, 40, 35, 30],
    'Sales': [20000, 40000, 60000, 80000, 100000]
}

sales_df = pd.DataFrame(sales_data)
print(sales_df)
```

```python
from sklearn.linear_model import LinearRegression
```

### 1. Student data

```python
X = student_df[["Hours_Studied","Attendance"]]
y = student_df["Marks"]
model = LinearRegression()
model.fit(X, y)
```

```python
predicted_marks = model.predict([[7,65]])
print("Predicted Marks for 7 Hours:", predicted_marks[0])
```

### 2. Salary data

```python
# Train salary model
X_salary = salary_df[["Experience"]]
y_salary = salary_df["Salary"]
salary_model = LinearRegression()
salary_model.fit(X_salary, y_salary)
print("Predicted Salary (6 years exp):", salary_model.predict([[6]])[0])
```

Feature & Label Identification

| Dataset         | Feature(s) (X)                                | Label (y) |
| --------------- | --------------------------------------------- | --------- |
| Student Dataset | Hours_Studied, Attendance                     | Marks     |
| Salary Dataset  | Experience                                    | Salary    |
| Sales Dataset   | Advertising_Spend, Salespeople, Product_Price | Sales     |



### 3. Sales data

```python
X_sales = sales_df[['Advertising_Spend']]
y_sales = sales_df['Sales']
sales_model = LinearRegression()
sales_model.fit(X_sales, y_sales)

print("Predicted Sales (6000 spend):", sales_model.predict([[6000]])[0])
```

### Task 6: Real-World ML Examples 
1. House price prediction 
2. Weather prediction 
3. Employee salary prediction 

 

#### 1. House price prediction

```python
house_df = pd.DataFrame({
    'Area': [1000, 1200, 1500, 1800, 2000],
    'Price': [3000000, 3500000, 4500000, 5500000, 6000000]
})

X = house_df[['Area']]
y = house_df['Price']

model = LinearRegression()
model.fit(X, y)
```

```python
predicted_price = model.predict([[1600]])

print("Predicted House Price:", predicted_price[0])
```

#### 2. Weather Prediction

```python
weather_df = pd.DataFrame({
    'Temperature': [25, 27, 30, 32, 35],
    'Humidity': [60, 65, 70, 75, 80],
    'Rainfall': [5, 10, 20, 30, 40]
})

X = weather_df[['Temperature', 'Humidity']]
y = weather_df['Rainfall']

model = LinearRegression()
model.fit(X, y)
```

```python
predicted_rainfall = model.predict([[28, 68]])

print("Predicted Rainfall:", predicted_rainfall[0])
```

#### 3. Employee Salary Prediction

```python
salary_df = pd.DataFrame({
    'Experience': [1, 2, 3, 4, 5],
    'Salary': [25000, 30000, 40000, 50000, 60000]
})

X = salary_df[['Experience']]
y = salary_df['Salary']

model = LinearRegression()
model.fit(X, y)

```

```python
predicted_salary = model.predict([[6]])

print("Predicted Salary:", predicted_salary[0])
```
