### Task 4 
Research YOLOv3, YOLOv5 and YOLOv8. 

#### YOLOv3 (2018)

- Released by Joseph Redmon in 2018.
- Uses the **Darknet-53** backbone with residual connections.
- Performs **multi-scale detection**, allowing it to detect small, medium, and large objects.
- One of the first YOLO versions to achieve strong real-time performance.

**Advantages:**
- Fast object detection.
- Good balance between speed and accuracy.
- Effective for real-time applications.

**Limitations:**
- Lower accuracy than newer YOLO versions.
- More difficult to train and deploy than PyTorch-based models.

---

#### YOLOv5 (2020)

- Released by Ultralytics in 2020.
- Implemented in **PyTorch**, making training and deployment easier.
- Available in multiple sizes: **Nano (n), Small (s), Medium (m), Large (l), and XLarge (x)**.
- Introduced features such as:
  - Auto-learning bounding boxes
  - Mosaic data augmentation
  - Easy export to ONNX, TensorRT, and TFLite

**Advantages:**
- Faster training and inference.
- Easy to use and customize.
- Strong community support.
- Widely adopted in industry.

**Limitations:**
- Newer versions provide better performance and more features.

---

#### YOLOv8 (2023)

- Released by Ultralytics in 2023.
- Uses an improved **anchor-free detection architecture**.
- Supports multiple computer vision tasks in one framework:
  - Object Detection
  - Image Classification
  - Instance Segmentation
  - Pose Estimation
  - Object Tracking
- Optimized for edge devices and real-time applications.

**Advantages:**
- Higher accuracy and speed than YOLOv5.
- Easier training and deployment.
- Better customization and scalability.
- State-of-the-art performance for many applications.

**Limitations:**
- Larger models may require more computational resources.
