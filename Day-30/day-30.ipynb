### Task 2: Line Plot Practice 
1. Create simple line graph 
2. Add labels 
3. Add title 

```python
import  matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.show()
```

### Task 3: Styling Graphs 
1. Add marker 
2. Change line style 
3. Add grid 

```python
#Changing Line Style 
plt.plot(x, y, linestyle="--", marker="o") 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Styled Line Graph")
plt.grid(True)
plt.show() 
```

### Task 4: Bar Chart Practice 
1. Student marks comparison 
2. Sales comparison 
3. Product comparison 

```python
#student mark comparison
students = ["Alok","Divya","Nathan","Nithya","Sam"]
marks = [67,78,90,75,82]
plt.bar(students,marks)
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student Marks Comparison")
plt.xticks(rotation=45)
plt.show()
```

```python
#sales comparison
year =["2019","2020","2021","2022","2023"]
sales= [34000,56000,45000,67000,79000]
plt.bar(year,sales)
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.xticks(rotation=45)
plt.show()
```

```python
#product comparison
products =["Laptop","Tablet","Smartphone","Smart Watch"]
units= [90,78,85,67]
plt.bar(products,units)
plt.xlabel("Products")
plt.ylabel("Units")
plt.title("Product comparison")
plt.xticks(rotation=45)
plt.show()
```

### Task 5: Pie Chart Practice 
1. Department distribution 
2. Expense analysis 
3. Course popularity 

```python
# Department distribution
departments = ["HR", "IT", "Finance", "Sales"]
employees = [50, 60, 45, 25]
plt.pie(employees, labels=departments, autopct='%1.1f%%',startangle=90)
plt.title("Department Distribution")
plt.show()
```

```python
# Expense analysis
expenses = ["Rent", "Food", "Transport", "Entertainment"]
values = [400, 300, 200, 100]
plt.pie(values, labels=expenses, autopct='%1.1f%%', startangle=90)
plt.title("Expense Analysis")
plt.show()
```

```python
#course popularity
course = ["DSA","FSD","DevOps","AI&ML","Power BI"]
enrol = [35,47,40,50,10]
plt.pie(enrol, labels=course, autopct="%1.1f%%", startangle=90)
plt.title("Course Popularity")
plt.show()
```

### Task 6: Scatter Plot Practice 
1. Height vs weight 
2. Study hours vs marks 
3. Temperature vs sales

```python
height =[155,149,153,165,160,161,166,170,163,175]
weight = [55,45,50,52,59,51,54,65,56,72]
plt.scatter(height,weight)
plt.show()
```

```python
# Study hours vs marks
study_hours = [1, 2, 3, 4, 5, 6, 7]
marks = [45, 50, 60, 65, 70, 85, 90]
plt.scatter(study_hours, marks, color='green')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks (%)")
plt.show()
```

```python
# Temperature vs sales
temperature = [20, 25, 30, 35, 40]
sales = [100, 200, 300, 400, 500]
plt.scatter(temperature, sales, color='red')
plt.title("Temperature vs Sales")
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Sales")
plt.show()
```
