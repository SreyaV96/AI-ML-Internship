### Task 3 
Lemmatize: 
Cars 
Children 
Mice 
Dogs 

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
words = ["Cars", "Children", "Mice", "Dogs"]

for word in words:
    print(lemmatizer.lemmatize(word.lower(), pos="n"))
```

### Task 4 
Compare outputs of: 
- Stemming 
- Lemmatization 


for the same words. 

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ["carrying", "crying", "studied", "children", "books", "mythologies"]

print("{:<15} {:<15} {:<15}".format("Word", "Stemmed", "Lemmatized"))
print("-" * 45)

for word in words:
    print("{:<15} {:<15} {:<15}".format(
        word,
        stemmer.stem(word),
        lemmatizer.lemmatize(word)
    ))
```

### Task 5 
Create a table containing: 

Word Stemmed Lemmatized 

for 10 words. 

 

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
import pandas as pd

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = [
    ("Crying", "v"),
    ("Working", "v"),
    ("Learning", "v"),
    ("Running", "v"),
    ("Cars", "n"),
    ("Children", "n"),
    ("Mice", "n"),
    ("Carrying", "v"),
    ("Studies", "v"),
    ("Mythologies", "n")
]

data = []

for word, pos in words:
    data.append([
        word,
        stemmer.stem(word.lower()),
        lemmatizer.lemmatize(word.lower(), pos=pos)
    ])

df = pd.DataFrame(data, columns=["Word", "Stemmed", "Lemmatized"])

print(df)
```
