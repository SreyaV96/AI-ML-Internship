### Task 2: Math Practice 
1. Find square root 
2. Find power 
3. Round numbers 
4. Use π value 

```python
import math
#square root
print("Square root of 625: ", math.sqrt(625))
#power
print("3 power 3:", math.pow(3,3))
#round numbers
print("round up: ", math.ceil(4.66))
print("round down: ",math.floor(4.66))
#use of pi
print("Value of pi:", math.pi)
```

### Task 3: Random Practice 
1. Generate random number 
2. Pick random element 
3. Generate list of random numbers 

```python
import random
#genearte random number
print("random number:",random.randint(1,20))

#pick random element
names =["Merlin", "Hridya", "Anjana", "Neethu", "Meghna", "Meera"]
print("random element:", random.choice(names))

#generate list of random numbers
marks = []
for i in range (1,11) :
    marks.append(random.randint(1,100))
print("marks: ", marks)
    
```

```python
random_list = [random.randint(1, 50) for _ in range(5)]
print("Random list:", random_list)
```

### Task 4: Datetime Practice 
1. Print current date 
2. Format date 
3. Extract year, month 

```python
import datetime

# current date
now = datetime.datetime.now()
print("Current date time:", now)

# format date
format_date = now.strftime("%Y-%m-%d")
print("Formatted date:", format_date)

# extract year, month
print("Year:", now.year)
print("Month:", now.month)

```

### Task 5: Visualization 
1. Create line graph 
2. Create bar chart 
3. Plot student marks 

```python
import matplotlib.pyplot as plt
#create line chart
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line chart")
plt.show()
```

```python
#create bar chart
students = ["A", "B", "C", "D"]
marks = [65, 87, 56, 90]
plt.bar(students, marks)
plt.title("Bar Chart ")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#plot student marks
plt.plot(students, marks, marker='o')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance")
plt.show()
```
