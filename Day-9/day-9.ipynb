### 2. Read and store in list of dictionary 

```python
with open ("students.txt", "r") as f:
    print(f.read())
```

```python
students = []
with open("students.txt", "r") as f:
    for line in f :
        name, marks =line.strip().split(", ")
        students.append({"name":name, "marks":int(marks)})
print(students)
```

### 3. Print all students 

```python
for student in students:
    print(student)
```

### 4. Find average marks 

```python
avg = sum(s["marks"] for s in students)/len(students)
print(avg)
```

### 5. Find topper 

```python
topper = max(students, key=lambda x: x["marks"])
print("Topper:", topper)
```

```python
# Pass mark = 45
passed = len([s for s in students if s["marks"] >= 45])
failed = len([s for s in students if s["marks"] < 45])

print("Passed Students:", passed)
print("Failed Students:", failed)
```
