## Task 2: Data Practice 
1. Create list of marks 
2. Find average 
3. Find max and min 
4. Count total students

```python
marks  = [67,45,56,78,90,89,77,96,59]
avg = sum(marks)/len(marks)
print("Average: ",avg)
print("Max: ", max(marks))
print("Min: ", min(marks))
print("Total students: ", len(marks))
```

## Task 3: Data Cleaning 
1. Remove None values 
2. Remove negative values 
3. Clean invalid data 

```python
data = [34,67,None,89,78,None,45,76,78,-45,90,-66,98, 120,113]
#Remove None values
clean_1 = [x for x in data if x is not None]
print(clean_1)

#Remove negative values
clean_2 = [x for x in  clean_1 if x>=0]
print(clean_2)

#clean invalid data
cleaned_data = [x for x in clean_2 if x<=100]
print(cleaned_data)
```

## Task 4: Data Filtering 
1. Find passed students 
2. Filter marks > 75 
3. Filter even numbers 

```python
#passed students
passed = [ x for x in cleaned_data if x>=45]
print("Passed: ", passed)

#filter marks >75
top_marks =[x for x in cleaned_data if x>75]
print("Top marks: ", top_marks)

#filter even numbers
even = [x for x in cleaned_data if x%2==0]
print("Even numbers: ", even)
```

## Task 5: Data Transformation 
1. Convert marks to percentage 
2. Multiply values 
3. Normalize data 

```python
# Convert to percentage scale
scaled = [x / 100 for x in cleaned_data]
print("Scaled:", scaled)

# Multiply values
multiplied = [x * 2 for x in cleaned_data]
print("Multiplied:", multiplied)

# Normalize data
max_val = max(cleaned_data)
min_val = min(cleaned_data)

normalized = [(x - min_val) / (max_val - min_val) for x in cleaned_data]
print("Normalized:", normalized)
```
