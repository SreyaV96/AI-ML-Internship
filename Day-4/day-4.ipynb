## Task 3: Exception Handling Practice 


### 1. Handle divide by zero 

```python
try: 
    x = int(input("Enter number: ")) 
    print(10 / x) 
except ZeroDivisionError: 
    print("Cannot divide by zero") 
```

### 2. Handle invalid input

```python
try: 
    x = int(input("Enter number: ")) 
    print(10 / x) 
except ZeroDivisionError: 
    print("Cannot divide by zero") 
except ValueError:
    print("invalid input")
```

```python
try:
    num = int(input("Enter a number: "))
    print("You entered:", num)
except ValueError:
    print("Error: Invalid input! Please enter a valid number")
```

### 3. Use try-except-else-finally 

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Result:", result)
finally:
    print("Execution completed")
```

## Task 4: Logic Challenge 


### 1. Find longest word in file 

```python
with open ("task2.txt", "r") as f:
    text = f.read()
    words = text.split()
    print("longest word :", max(words, key= len))
```

### 2. Count vowels in file 

```python
with open ("task2.txt","r") as f:
    content  = f.read().lower()
    vowels = "aeiou"
    count = 0
    for char in content :
        if char in vowels :
            count += 1
    print("Number of vowels:", count)
```

## 3. Remove duplicate lines 

```python
unique_lines = set()

with open("task2.txt", "r") as f:
    for line in f:
        unique_lines.add(line)

with open("output.txt", "w") as f:
    for line in unique_lines:
        f.write(line)
```
