### Task 2 
Create TF-IDF vectors for: 
- AI is amazing 
- AI is powerful 
- Machine learning is amazing

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
documents = [
    "AI is amazing", 
    "AI is powerful", 
    "Machine learning is amazing"]
X = vectorizer.fit_transform(documents)
print(X.toarray())
```

### Task 3 
Print vocabulary using: 
get_feature_names_out() 

```python
print(vectorizer.get_feature_names_out())
```

### Task 4 
Compare BoW and TF-IDF outputs. 

```python
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer()
X = bow.fit_transform(documents)
print (X.toarray())
```

```python
print(bow.get_feature_names_out())
```

##### Difference

- BoW stores only word counts.
- TF-IDF stores weighted values based on word importance.


### Task 5 
Identify the most important words using TF-IDF scores. 

```python
import pandas as pd
```

```python
documents = [
    "AI is amazing",
    "AI is powerful",
    "Machine learning is amazing"
]

tfidf = TfidfVectorizer()

matrix = tfidf.fit_transform(documents)

words = tfidf.get_feature_names_out()

df = pd.DataFrame(
    matrix.toarray(),
    columns=words,
    index=["Doc1", "Doc2", "Doc3"]
)

print(df)

print("\nHighest TF-IDF word in each document:")

for i, row in df.iterrows():
    print(i, ":", row.idxmax(), "=", round(row.max(), 3))
```
