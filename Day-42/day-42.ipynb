```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
import warnings
warnings.filterwarnings('ignore')
```

##### 1. Student Dataset

```python
#create dataset
data = {
    "Hours": [2, 3, 5, 6, 1, 4, 7, 8, 2, 6],
    "Attendance": [60, 70, 90, 85, 50, 75, 95, 98, 65, 88],
    "Assignments": [2, 3, 5, 4, 1, 4, 5, 5, 2, 4],
    "Result": [0, 0, 1, 1, 0, 1, 1, 1, 0, 1]}

student_df = pd.DataFrame(data)

X_student = student_df[["Hours","Attendance","Assignments"]]
y_student = student_df["Result"]
#create model
model_s = GaussianNB()
#train model
model_s.fit(X_student,y_student)
#prediction
print(model_s.predict([[3.5,75,4]]))
```

##### 2. Disease dataset

```python
#create dataset
data = {
    "Age": [25, 30, 45, 50, 35, 60, 55, 40, 28, 65],
    "BMI": [22, 24, 31, 34, 26, 36, 33, 29, 23, 38],
    "BP": [110, 115, 145, 150, 120, 160, 155, 135, 118, 170],
    "Disease": [0, 0, 1, 1, 0, 1, 1, 1, 0, 1]
}
disease = pd.DataFrame(data)

X_disease = disease[["Age","BMI","BP"]]
y_disease = disease["Disease"]
#create model
model_d = GaussianNB()
#train model
model_d.fit(X_disease,y_disease)
#prediction
print(model_d.predict([[42,26,150]]))
```

##### 3. Spam dataset

```python
#create dataset
data = {
    "EmailLength": [50, 80, 120, 40, 150, 90, 200, 110, 60, 170],
    "Links": [5, 6, 1, 7, 0, 4, 0, 2, 5, 1],
    "UrgentWords": [4, 5, 0, 6, 0, 3, 0, 1, 4, 0],
    "Attachments": [2, 3, 0, 1, 0, 2, 0, 1, 2, 0],
    "Spam": [1, 1, 0, 1, 0, 1, 0, 0, 1, 0]
}
spam_df = pd.DataFrame(data)

X = spam_df[["EmailLength", "Links", "UrgentWords", "Attachments"]]
y = spam_df["Spam"]
#create model
model = GaussianNB()
#train model
model.fit(X, y)
#prediction
print(model.predict([[75, 4, 3, 1]]))
```

### Task 5: Probability Activity 
Suppose: 
- Total Emails = 200 
- Spam Emails = 150 
- Calculate: 
    1. Probability of Spam 
    2. Probability of Not Spam 

```python
total_emails = 200
spam_emails = 150
p_spam = spam_emails/total_emails
print("Probability of spam:", p_spam)
p_notSpam = 1- p_spam
print("Probability of Not spam:", p_notSpam)
```
