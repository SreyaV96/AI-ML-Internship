### Task 2 
Create a small dataset with 5 sentences. 

```python
sentences = [
    "I love machine learning",
    "Machine learning is interesting",
    "Python is used for machine learning",
    "I love Python programming",
    "Artificial intelligence uses machine learning"]
```

### Task 3 
Train a Word2Vec model. 

```python
from gensim.models import Word2Vec
```

```python
# tokenize
data = [sentence.lower().split() for sentence in sentences]
print (data)
```

```python
model = Word2Vec(sentences=data,
                 vector_size =50,
                 window =3,
                 min_count =1)
```

### Task 4 
Print word vectors. 

```python
print("Vector for 'machine':")
print(model.wv['machine'])

print("\nVector for 'python':")
print(model.wv['python'])
```

### Task 5 
Find similar words using: 

most_similar() 

```python
similar_words= model.wv.most_similar("python") 
print(similar_words)
```

```python
print("Words similar to 'python':")
for word, score in similar_words:
    print(word, ":", score)
```
