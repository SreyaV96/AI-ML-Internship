### Task 5 
Compare Rule-Based and AI-Based chatbots in a table. 

# Rule-Based vs AI-Based Chatbots (Simple)

| Point | Rule-Based Chatbot | AI-Based Chatbot |
|---|---|---|
| **What it is** | Follows fixed rules set by humans | Learns from data and understands language |
| **How it answers** | Matches exact words/keywords | Understands meaning, even if words are different |
| **Flexibility** | Can only answer what it's programmed for | Can answer new or unexpected questions |
| **Learning** | Cannot learn on its own | Improves with more data/training |
| **Conversation style** | Robotic, script-based | Natural, human-like |
| **Setup difficulty** | Easy and quick to build | Harder, needs training and data |
| **Cost** | Low cost | Higher cost |
| **Example** | FAQ bots, basic customer service bots | ChatGPT, Siri, Alexa |
