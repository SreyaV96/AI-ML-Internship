## Task 4: Refactoring 
  Convert: 
   1. Long code → functions 

   2. Repeated logic → reusable function 

```python
# Long code → functions
def calculate_area(l, b):
    return l * b

def calculate_perimeter(l, b):
    return 2 * (l + b)

print("Area:", calculate_area(2, 3))
print("Perimeter:", calculate_perimeter(2, 3))
```

```python
#Repeated logic → reusable function 
def is_even(num):
    return num % 2 == 0

print(is_even(4))
print(is_even(7))
print(is_even(10))
```
