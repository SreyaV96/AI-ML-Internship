```python
#create dataset
import pandas as pd
import numpy as np
import random

random.seed(42)
np.random.seed(42)

courses = ["FSD","DSA","DevOps","Power BI","AI/ML","UI/UX"]
cities = ["CLT","TVM","EKM"]
years = [2019, 2020, 2021, 2022, 2023, 2024]

data_list = []

for i in range(500):
    data_list.append({
        "Enroll_ID": 10001 + i, 
        "Course": random.choice(courses),
        "Year": random.choice(years),
        "City": random.choice(cities),
        "Marks": random.randint(150, 200),
        "Placement": random.choice(["Yes","No"])
    })

df = pd.DataFrame(data_list)

print(df.head())
print(df.shape)
```

```python
#create 4 subplot
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# 1. Marks Distribution
plt.subplot(2,2,1)
plt.hist(df["Marks"], bins=20)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution")

# 2. Enrollment by Year
plt.subplot(2,2,2)
year_counts = df["Year"].value_counts().sort_index()
plt.bar(year_counts.index, year_counts.values)
plt.title("Enrollment by Year")
plt.xlabel("Year")
plt.ylabel("Number of Enrollments")

# 3. Course Popularity
plt.subplot(2,2,3)
course_counts = df["Course"].value_counts()
plt.bar(course_counts.index, course_counts.values)
plt.title("Course Popularity")
plt.xlabel("Course Name")
plt.ylabel("Number of Enrollments")
plt.xticks(rotation=45)

# 4. City-wise Enrollment 
plt.subplot(2,2,4)
city_counts = df["City"].value_counts()
plt.pie(city_counts.values, labels=city_counts.index, autopct='%1.1f%%')
plt.title("City-wise Enrollment")

plt.tight_layout()
plt.show()
```
