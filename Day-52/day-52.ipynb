### Task 3 
Remove punctuation from: 
Python!!! is Awesome??? 

```python
import string
text= "Python!!! is Awesome???"
clean_text = text.translate(str.maketrans('', '', string.punctuation))
print(clean_text)
```

### Task 5 
Clean this sentence: 
AI 2025 is AMAZING!!!   

```python
import re
text = "Hello!!! AI 2025 is Amazing." 
text = text.lower() 
text = re.sub(r'[^a-zA-Z\s]', '', text) 
print(text) 
```
