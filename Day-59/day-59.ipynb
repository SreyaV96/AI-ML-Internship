### Task 1 
Create a chatbot that responds to: 
- Hello 
-  Hi 
-  Bye 
- Your Name 

```python
print("AI Assistant") 
print("Type bye to exit.\n")
while True:
    user = input ("You:").lower()
    if user == "hello":
        print("Bot: Hello! How can I help you?")
    elif user == "hi":
        print("Bot: Hi! How can I help you?")
    elif user == "your name":
        print("Bot: I am an AI chatbot")
    elif user == "bye":
        print("Bot: Goodbye!")
        break
    else:
        print("Bot: Sorry, I don't understand.Can you please rephrase")
```

### Task 2 
Create a College Enquiry Chatbot with at least five questions. 

```python
print("===== College Enquiry AI Assistant =====")
print("Type 'bye' or 'thankyou' to exit.\n")

while True:
    user = input("You: ").lower()

    if user == "hi":
        print("Bot: Hi! Welcome to our College Enquiry Assistant. How can I help you today?")

    elif user == "hello":
        print("Bot: Hello! Welcome to our College Enquiry Assistant. How may I assist you?")

    elif user == "your name":
        print("Bot: I am the College Enquiry AI Assistant.")

    elif user == "location":
        print("Bot: Our college is located in Kochi, Kerala.")

    elif user == "courses offered in computer science":
        print("Bot: We offer B.Sc. Computer Science, BCA, M.Sc. Computer Science, and MCA.")

    elif user == "university":
        print("Bot: Our college is affiliated with Mahatma Gandhi (MG) University.")

    elif user == "course fee for b.sc. computer science":
        print("Bot: The course fee for B.Sc. Computer Science is ₹25,000 per semester.")

    elif user == "when will admissions open":
        print("Bot: Admissions are expected to begin in July 2026.")

    elif user == "how can i apply for a management quota seat":
        print("Bot: To apply for a management quota seat, please contact our admission office at 9776543110.")

    elif user in ("thankyou","thank you"):
        print("Bot: You're welcome! Thank you for contacting us. Have a great day!")
        break

    elif user == "bye":
        print("Bot: Goodbye! Thank you for contacting us. We wish you all the best.")
        break

    else:
        print("Bot: Sorry, I couldn't understand your question. Please contact our office for further assistance.")
```

### Task 5 
Add a default response for unknown questions.

Example: Sorry, I don't understand your question. 

```python
print("===== College Enquiry AI Assistant =====")
print("Type 'bye' or 'thankyou' to exit.\n")
responses = {"hi": "Hi! Welcome to our College Enquiry Assistant. How can I help you today?",
             "hello": "Hello! Welcome to our College Enquiry Assistant. How may I assist you?",
             "your name": "I am the College Enquiry AI Assistant.",
             "location": "Our college is located in Kochi, Kerala.",
             "courses offered in computer science": "We offer B.Sc. Computer Science, BCA, M.Sc. Computer Science, and MCA.",
             "university": "Our college is affiliated with Mahatma Gandhi (MG) University.",
             "admission": "Admissions are expected to begin in July 2026.",
             "contact":"Call us at 9876543210.",
             "management quota": "To apply for a management quota seat, please contact our admission office at 9776543110.",
             "course fee for b.sc. computer science": "The course fee for B.Sc. Computer Science is ₹25,000 per semester.",
             }

while True :
    user = input("You:").lower()
    if user == "bye":
        print("Bot: Goodbye! Thank you for contacting us. We wish you all the best.")
        break
    elif user in ("thankyou", "thank you"):
        print("Bot: You're welcome! Thank you for contacting us. Have a great day!")
        break
    found = False
    for key in responses:
        if key in user :
            print("Bot:",responses[key])
            found = True
            break 
    if not found:
        print("Bot: Sorry, I couldn't understand your question. Please contact our office for further assistance")
    
```
