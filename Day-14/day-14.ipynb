## Task 2: JSON Practice 
1. Convert dictionary to JSON 

2. Convert JSON to dictionary 
3. Write JSON to file 
4. Read JSON from file 

```python
#convert dict to JSON
import json
dict_1 = {"name": "Sravya", "course":"BCA", "CGPA": 8.6}
json_1 = json.dumps(dict_1)
print(json_1)
```

```python
#convert JSON to dict
dict_1 = json.loads(json_1)
print(dict_1)
print(type(dict_1))

```

```python
#write JSON to file
with open ("data.json", "w") as f:
    json.dump(dict_1,f)
```

```python
#read JSON from a file
with open ("data.json","r") as f:
    data = json.load(f)
print(data)
```

## Task 3: CSV Practice 
1. Create CSV file 
2. Write student data 
3. Read CSV file 
4. Print formatted output 

```python
#create a csv file and write
import csv
with open("students.csv","w",newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Id", "Name","Marks", "Grade"])
    writer.writerow([1, "Anagha", 65, "C"])
    writer.writerow([2, "Rahul", 75, "B"])
    writer.writerow([3, "HariKrishnan", 81, "A"])
    writer.writerow([4, "Sravya", 90, "A+"])
print("CSV file created successfully!")   
```

```python
#Read CSV file and print formatted output
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)  # Skip header

    print("\nStudent Details:")
    for row in reader:
        print(f"ID: {row[0]}, Name: {row[1]}, Marks: {row[2]}, Grade: {row[3]}")
```

## Task 4: Combined Task 
  Program: 
   - Read CSV file 
   - Convert to dictionary 
   - Save as JSON 

```python

data = []

# Read CSV file
with open("students.csv", "r") as file:
    reader = csv.DictReader(file)

    # Convert rows to dictionary list
    for row in reader:
        data.append(row)

# Save as JSON
with open("students.json", "w") as json_file:
    json.dump(data, json_file, indent=4)

print("CSV converted to JSON successfully!")
```
