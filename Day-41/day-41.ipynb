```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings ('ignore')
```

##### 1. Student dataset

```python
#create dataset
data = {
    "Hours": [2, 3, 5, 6, 1, 4, 7, 8, 2, 6],
    "Attendance": [60, 70, 90, 85, 50, 75, 95, 98, 65, 88],
    "Assignments": [2, 3, 5, 4, 1, 4, 5, 5, 2, 4],
    "Result": [0, 0, 1, 1, 0, 1, 1, 1, 0, 1]}

student_df = pd.DataFrame(data)

X_student = student_df[["Hours","Attendance","Assignments"]]
y_student = student_df["Result"]
#create model
model_s = RandomForestClassifier(random_state=43)
#train model
model_s.fit(X_student,y_student)
#prediction
print(model_s.predict([[3.5,75,4]]))
```

```python
model_s10 = RandomForestClassifier(n_estimators=10, random_state=43)
model_s10.fit(X_student,y_student)
print("10 trees:",model_s10.predict([[3.5,75,4]]))

model_s50 = RandomForestClassifier(n_estimators=50, random_state=43)
model_s50.fit(X_student,y_student)
print("50 trees:",model_s50.predict([[3.5,75,4]]))

model_s100 = RandomForestClassifier(n_estimators=100, random_state=43)
model_s100.fit(X_student,y_student)
print("100 trees:",model_s100.predict([[3.5,75,4]]))
```

##### 2. Disease dataset


```python
#create dataset
data = {
    "Age": [25, 30, 45, 50, 35, 60, 55, 40, 28, 65],
    "BMI": [22, 24, 31, 34, 26, 36, 33, 29, 23, 38],
    "BP": [110, 115, 145, 150, 120, 160, 155, 135, 118, 170],
    "Disease": [0, 0, 1, 1, 0, 1, 1, 1, 0, 1]
}
disease = pd.DataFrame(data)

X_disease = disease[["Age","BMI","BP"]]
y_disease = disease["Disease"]
#create model
model_d = RandomForestClassifier(random_state=43)
#train model
model_d.fit(X_disease,y_disease)
#prediction
print(model_d.predict([[42,26,150]]))
```

```python
model_d10 = RandomForestClassifier(n_estimators=10, random_state=43)
model_d10.fit(X_disease,y_disease)
print("10 trees:",model_d10.predict([[42,26,150]]))

model_d50 = RandomForestClassifier(n_estimators=50, random_state=43)
model_d50.fit(X_disease,y_disease)
print("50 trees:",model_d50.predict([[42,26,150]]))

model_d100 = RandomForestClassifier(n_estimators=100, random_state=43)
model_d100.fit(X_disease,y_disease)
print("100 trees:",model_d100.predict([[42,26,150]]))
```

##### 3. Loan approval dataset

```python
#create dataset
data = {
    "Income": [25000, 30000, 45000, 50000, 70000, 80000, 35000, 60000, 90000, 40000],
    "CreditScore": [580, 620, 700, 720, 780, 800, 650, 750, 820, 680],
    "Status": [0, 0, 1, 1, 1, 1, 0, 1, 1, 1]}
loan = pd.DataFrame(data)

X_loan = loan[["Income","CreditScore"]]
y_loan = loan ["Status"]
#create model
model_l = RandomForestClassifier(random_state=43)
#train model
model_l.fit(X_loan,y_loan)
#prdeiction
print(model_l.predict([[22000,690]]))

```

```python
model_l10 = RandomForestClassifier(n_estimators=10, random_state=43)
model_l10.fit(X_loan,y_loan)
print("10 trees:",model_l10.predict([[22000,690]]))

model_l50 = RandomForestClassifier(n_estimators=50, random_state=43)
model_l50.fit(X_loan,y_loan)
print("50 trees:",model_l50.predict([[22000,690]]))

model_l100 = RandomForestClassifier(n_estimators=100, random_state=43)
model_l100.fit(X_loan,y_loan)
print("100 trees:",model_l100.predict([[22000,690]]))
```
