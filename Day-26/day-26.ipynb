### Task 2: Series Practice 
1. Create Series 
2. Create Series with custom index 
3. Access Series values

```python
import pandas as pd
#create series
a = pd.Series([10,30,50,70,90])
print(a)
```

```python
#create series with custom index
marks = pd.Series([59,78,67,80], index=["Maths", "Science", "Biology", "English"])
print (marks)
# access series values
marks["English"]
```

### Task 3: DataFrame Practice 
1. Create DataFrame using dictionary 
2. Print DataFrame 
3. Add one more column 

```python
students = {"Name":["Aparna","Arjun","Megha","Meenu","Mithun","Neethu","Prince","Sourav"],
            "Marks":[67,72,56,60,89,59,73,80]}
df = pd.DataFrame(students)
print(df)
#add new column 
df["Grade"] = ["B","B+","C","C+","A","C","B+","A"]
print("Students df:\n",df)
```
