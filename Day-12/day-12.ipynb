## Task 2: List Comprehension 
1. Square numbers 
2. Find even numbers 
3. Convert strings to uppercase 
4. Filter numbers > 50 

```python
# Square numbers
num = [i for i in range (1,11)]
squares = [x*x for x in num]
print (squares)
```

```python
# even numbers
even = [x for x in num if x%2==0]
print(even)
```

```python
# Convert strings to uppercase 
name = ["Aparna", "Veena", "Sarang", "Neil"]
name_upper = [word.upper() for word in name]
print(name_upper)
```

```python
# Filter numbers > 50 
num = [34,78,56,23,68,45,99,100,86,63,55]
n_1 = [x for x in num if x>50]
print(n_1)
```

## Task 3: Dictionary Comprehension 
1. Create number-square dictionary 
2. Filter even numbers 
3. Map names to lengths 

```python
#Create number-square dictionary 
num =[2,4,6,8,10]
num_2 ={x:x*x for x in num}
print (num_2)
```

```python
#Filter even numbers
even_num = {x:y for x,y in num_2.items() if x%2==0}
print(even_num) 
```

```python
# Map names to lengths 
names = ["Anagha", "Sreelekshmi" ,"Samson", "Dhanush"]
name_len ={name : len(name) for name in names}
print(name_len)
```

## Task 5: Generator Practice 
1. Create generator for numbers 
2. Generate even numbers 
3. Fibonacci using generator 

```python
# Create generator for numbers

def numbers(n):
    for i in range(n):
        yield i

for num in numbers(5):
    print(num)
```

```python
# Generate even numbers 
def even_num(n):
    for i in range(n):
        if i%2==0:
            yield i

for num in even_num(10):
    print(num)
```

```python
# Fibonacci using generator 
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(10):
    print(num)
```

## Task 6: Zip & Enumerate 
1. Combine 2 lists 
2. Print index + value 
3. Convert zip result to dictionary 

```python
# Combine 2 lists 
names = ["Arathi", "Anjana", "Kavya", "Meera", "Sonika"]
marks = [45, 56, 89, 67, 72]
result = list(zip(names, marks))
print(result)
```

```python
# Print index + value 
for i, name in enumerate(names):
    print(i, name)
```

```python
# Convert zip result to dictionary 
result_dict = dict(zip(names, marks))
print(result_dict)
```
