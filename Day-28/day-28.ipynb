### Task 2: Missing Data Practice 
1. Create dataset with missing values 
2. Detect missing values 
3. Count missing values 

```python
import pandas as pd
import numpy as np
data = {"Name":["Amritha","Athira","Athul","Dhanush","Evon","Geethu","Gokul","Hari","Irin","Jancy"],
        "Marks":[167,180,175,190,np.nan,166,172,159,182,np.nan],
        "City":["Eranakulam","Trivandrum","Idukki",None,"Kollam","Wayanad","Calicut",None,"Kannur","Malappuram"]}
df = pd.DataFrame(data)
df_1 = df.copy()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

### Task 3: Remove Missing Data 
1. Drop rows with missing values 
2. Drop columns with missing values 

```python
#drop rows with missing values
df.dropna()
```

```python
#drop columns with missing values
df.dropna(axis=1)
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

### Task 4: Fill Missing Data 
1. Fill with text 
2. Fill with mean value 
3. Fill numeric columns 

```python
#fill with text
df_1["City"] = df_1["City"].fillna("Unknown")
```

```python
#fill with mean value
df_1["Marks"] = df_1["Marks"].fillna(df_1["Marks"].mean()).round(2)
```

```python
df_1
```

```python
#Fill numeric columns
numeric_cols = df_1.select_dtypes(include="number").columns
df_1[numeric_cols] = df_1[numeric_cols].fillna(df_1[numeric_cols].mean())
```

### Task 5: Duplicate Handling 
1. Create duplicate rows 
2. Detect duplicates 
3. Remove duplicates 

```python
#create duplicate rows
df_dup =pd.concat([df_1,df_1.iloc[:2]],ignore_index =True)
df_dup
```

```python
#detect duplicates
df_dup.duplicated()
```

```python
df_dup[df_dup.duplicated()]
```

```python
#remove duplicates
df_clean=df_dup.drop_duplicates()
df_clean
```
