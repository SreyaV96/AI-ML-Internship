### Task 2: Reshape Practice 
1. Convert 1D → 2D 
2. Convert 1D → 3D 
3. Try invalid reshape 

```python
import numpy as np
arr_1 = np.array([10,20,30,40,50,60])
print("2D array:", arr_1.reshape(2,3))
print("3D array:", arr_1.reshape(1,3,2))
```

### Task 3: Flatten 
1. Convert 2D → 1D 
2. Compare flatten vs ravel 

```python
#2D --> 1D
arr_2 = np.array([[2,4,6],
                  [8,10,12]])
print("1D array:", arr_2.flatten())
```

```python
print("ravel:", arr_2.ravel())
```

```python
#compare flatten vs ravel
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
flat = matrix.flatten()
print("\nFlattened Array:")
print(flat)
ravel_arr = matrix.ravel()


print("\nFlatten vs Ravel:")
print("flatten():", flat)
print("ravel():", ravel_arr)

# Modify original matrix
matrix[0, 0] = 100

print("\nAfter modifying original matrix:")
print("Original Matrix:")
print(matrix)

print("flatten() result:")
print(flat)      # unchanged copy

print("ravel() result:")
print(ravel_arr) # changed view
```

### Task 4: Transpose 
1. Transpose matrix 
2. Verify rows   columns 

```python
# transpose matrix
arr_3 = np.array([[1,2],[3,4],
                [5,6],[7,8]])
print("array:", arr_3)
print("transpose:", arr_3.T)
print("original shape:", arr_3.shape)
print("transposed shape:",arr_3.T.shape)
```

### Task 5: Stacking 
1. Vertical stack 
2. Horizontal stack 
3. Combine arrays 

```python
a = np.array([1,2])
b = np.array([6,7])
print("v stack:", np.vstack((a, b))) 
print("h stack:", np.hstack((a, b))) 
print("combined array:", np.concatenate((a,b)))
```
