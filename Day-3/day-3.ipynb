### 3. Mathematical Operations

```python
## Union
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

myset = set1.union(set2, set3, set4)
print(myset)
```

```python
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

myset = set1 | set2 | set3 |set4
print(myset)
```

```python
# intersection
a = {1, 2, 3}
b = {3, 4, 5}
set_1 = a&b
print(set_1)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}
set_1 = a.intersection(b)
print(set_1)
```

```python
print(a - b)
```

```python
#symmetric difference
print(a ^ b)
```
