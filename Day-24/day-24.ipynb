### Task 4: Random Choice 
1. Pick random color 
2. Generate multiple choices 
3. Simulate random names 

```python
color = ["red", "blue", "orange","green", "yellow", "violet", "black", "white", "brown"]
#pick random color
choice_1 = np.random.choice(color)
#generate multiple choices
choice_2 = np.random.choice(color, size=4)
choice_3 = np.random.choice(color, size= 7)
choice_4 = np.random.choice(color, size= 10)
print("choice 1:", choice_1)
print("choice 2:", choice_2)
print("choice 3:", choice_3)
print("choice 4:", choice_4)
```

```python
#simulate random names
names = ["Meera", "Anagha", "Sreehari", "Adithya","Geethu","Akhil","Aswin","Vishnu", "Sachin", "Sreya"]
list_1 = np.random.choice(names, size=5)
list_2 = np.random.choice(names, size=10)
print("list 1:", list_1)
print("list 2:", list_2)
```

### Task 5: Shuffle 
1. Shuffle array 
2. Shuffle student IDs 

```python
arr_1 = np.array([20,25,30,35,40,45,50,55,60])
print("array:", arr_1)
#shuffle array
np.random.shuffle(arr_1)
print("shuffled array:", arr_1)
```

```python
student_ids = np.array([101, 102, 103, 104, 105])
#shuffle student IDs
np.random.shuffle(student_ids)
print("Shuffled Student IDs:", student_ids)
```
