## Task 3: Nested Data Practice 

### 1. Create list of 5 students

```python
students = [{"Name":"Anand", "Marks":69},
            {"Name":"Sarah", "Marks":87},
            {"Name":"Meenu","Marks":45},
            {"Name":"Vishnu","Marks":58},
            {"Name":"Sreelakshmi","Marks":90}]
```

### 2. Print all students 

```python
for student in students:
    print(student)
```


### 3. Find average marks 

```python
total = 0
for student in students:
    total += student["Marks"]
avg = total/len(students)
print(avg)
```

### 4. Find highest marks 

```python
max_marks = max(student["Marks"] for student in students)
print(max_marks)
```

```python
highest = 0
for student in students:
    if student["Marks"] > highest:
        highest = student["Marks"]

print(highest)
```

### 5. Print students with marks > 75 

```python
marks_75 = [student["Name"] for student in students if student["Marks"]>75]
print(marks_75)
```

## Task 6: Logic Challenge 

### 1. Second highest number in list


```python
numbers = [30, 21, 89, 65, 85, 99]
numbers.sort()
print("Second highest Number is:",numbers[-2])
```

```python
def second_highest(lst):
    unique = list(set(lst)) 
    if len(unique) < 2:
        print("only one number")
    unique.sort()
    print( unique[-2])
```

```python
second_highest([1,2,8,5,3,2,9,19,12,4,9])
```

### 2. Count frequency of characters 

```python
def char_frequency(s):
    freq = {}
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq
```

```python
print(char_frequency("Hello World"))
```
