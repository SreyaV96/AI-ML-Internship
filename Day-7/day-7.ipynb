## Section C: Mini Task 
  Create program: 
 -  Take 5 student inputs 
 - Store in list 
 - Find average 
 - Find topper 

```python
students = [{"Name":"Anand", "Marks":69},
            {"Name":"Sarah", "Marks":87},
            {"Name":"Meenu","Marks":45},
            {"Name":"Vishnu","Marks":58},
            {"Name":"Sreelakshmi","Marks":90}]
```

```python
marks = [student["Marks"] for student in students]
average = sum(marks) / len(marks)
print(average)
```

```python
top = students[0] 
for s in students: 
    if s["Marks"] > top["Marks"]:
        top = s 
print(top) 
```
