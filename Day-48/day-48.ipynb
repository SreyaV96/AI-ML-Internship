### 1. LabelImg
- Open-source image annotation tool.
- Used for drawing bounding boxes manually.

#### Features:
- Easy to use.
- YOLO annotation format support.
- Pascal VOC support.
- Offline desktop application.

#### Best For:
Small and medium datasets.

### 2. Roboflow

Cloud-based dataset management and annotation platform.

#### Features:
- Annotation tools
- Dataset versioning
- Data augmentation
- YOLO export
- Team collaboration

#### Best For:
Professional computer vision projects.

### 3. CVAT

Advanced annotation platform developed by Intel.

#### Features:
- Bounding boxes
- Segmentation
- Tracking
- Team collaboration
- Large-scale datasets

#### Best For:

Enterprise-level annotation projects.

#### Helmet Detection Dataset Structure

The dataset follows the YOLO directory format:

```text
Helmet_detection_dataset/
│
├── images/
│   ├── train/
│   │   ├── img1.jpg
│   │   ├── img2.jpg
│   │   └── ...
│   │
│   └── val/
│       ├── val1.jpg
│       ├── val2.jpg
│       └── ...
│
├── labels/
│   ├── train/
│   │   ├── img1.txt
│   │   ├── img2.txt
│   │   └── ...
│   │
│   └── val/
│       ├── val1.txt
│       ├── val2.txt
│       └── ...
│
└── data.yaml
```

#### Description

* `images/train/` → Training images
* `images/val/` → Validation images
* `labels/train/` → YOLO annotations for training images
* `labels/val/` → YOLO annotations for validation images
* `data.yaml` → Dataset configuration file
