### Task 2: Marks Analysis 
1. Find average marks 
2. Find highest marks 
3. Find lowest marks 
4. Find topper 

```python
import numpy as np
marks =np.array([[75,64,80],
        [67,56,72],
        [87,72,70]])
print("average mark:", np.mean(marks))
print("highest marks:", np.max(marks))
print("lowest marks:", np.min(marks))
total = np.sum(marks, axis =1)
print("topper index:", np.argmax(total))
```

### Task 3: Image Operations 
1. Increase brightness 
2. Decrease brightness 
3. Normalize image pixels 

```python
image = np.array([[100,110,150],
                  [200,230,250],
                  [300,320,350]])
print("bright:", image+50)
print("dark:", image-50)
print("normalized:", image/np.max(image))
```

### Task 4: Dataset Filtering 
1. Filter values > 50 
2. Replace low values 
3. Count filtered values 

```python
arr_1 = np.array([12,18,9,45,34,82,35,45,78,56,69,98,32,37])
filtered = arr_1[arr_1>50]
print("filtered values:", filtered)
arr_1[arr_1<50]=0
print("replaced array:", arr_1)
print("count of values>50:",np.sum(arr_1>50))
```

### Task 5: Reshaping
