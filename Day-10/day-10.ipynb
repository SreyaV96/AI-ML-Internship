## Task 4: Method Practice 
### 1. Create method to: 
 - Check pass/fail 
 - Calculate grade 

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Check pass/fail
    def result(self):
        if self.marks >= 40:
            return "Pass"
        else:
            return "Fail"

    # Calculate grade
    def grade(self):
        if self.marks >= 90:
            return "A+"
        elif self.marks >= 75:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "F"

    # Print details
    def details(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
        print("Result:", self.result())
        print("Grade:", self.grade())
        print()
```

```python
s1 = Student("Neethu", 39)
s1.result()
```

```python
s1.grade()
```

```python
s1.details()
```

## Task 5: Object List 
### 1. Create list of students
### 2. Print all students 
### 3. Find topper 

```python
students = [
    Student("Rahul", 85),
    Student("Anu", 92),
    Student("Vishnu", 38)
]

# Print all students
for s in students:
    s.details()

```

```python
# Find topper
topper = max(students, key=lambda x: x.marks)
print("Topper:", topper.name)
print("Marks:", topper.marks)
```

##  Task 7: Logic Challenge 


### 1. Create class BankAccount 
 - deposit 
 - withdraw 
 - check balance 

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print("Balance:", self.balance)
```

```python
acc1 = BankAccount(5000)

acc1.deposit(2000)
acc1.withdraw(1000)
acc1.check_balance()
```

### 2. Create class Product 
 - name, price 
 - calculate discount 

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def calculate_discount(self, percent):
        discount = self.price * percent / 100
        final_price = self.price - discount
        print("Product Name:", self.name)
        print("Original Price:", self.price)
        print("Discount Price:", final_price)
```

```python
p1 = Product("Laptop", 50000)

p1.calculate_discount(10)
```
